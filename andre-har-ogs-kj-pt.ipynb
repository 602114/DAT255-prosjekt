```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)


import os
from fastai.vision.all import *
from fastai.collab import *
import cv2, matplotlib.pyplot as plt
```

```python
path = Path('../input/h-and-m-personalized-fashion-recommendations')
ar_df = pd.read_csv(path/'articles.csv')
tr_df = pd.read_csv(path/'transactions_train.csv')
```

```python
#Velger å utelukke kjøp gjort før en viss dato for å forkorte antall rader og forholde oss til gjeldene/nyere kjøp
ntr_df = tr_df[tr_df['t_dat'] > '2020-07-03']
```

```python
ntr_df.info()
```

```python
#Lager en dataframe med bare dato, kunde og artikkel
dca_df = ntr_df[['t_dat', 'customer_id', 'article_id']]

#Teller hvor mange ganger hver vare er kjøpt
antall = dca_df.article_id.value_counts()
dca_df.head()
```

```python
par = {}
for count,i in enumerate(antall.index.values[900:1058]):
    #Finner alle kundene som har kjøpt den valgte varen
    UnikeKunder = dca_df.loc[dca_df['article_id'] == i.item(),'customer_id'].unique()
    
    #Teller opp antall av hver vare som de samme kundene har kjøpt 
    antall2 = dca_df.loc[(dca_df['customer_id'].isin(UnikeKunder))&(dca_df['article_id'] != i.item()),'article_id'].value_counts()
    
    #Gjennom denne listen vil vi kunne se hvilke varer som er kjøpt flest ganger sammen med den valgte varen
    #Dermed kan vi finne de top 4 varene som ble kjøpt av samme kunde og legge disse inn i en liste som par 
    par[i.item()] = [antall2.index[0], antall2.index[1], antall2.index[2], antall2.index[3]]
```

```python
#Viser top 4 varer
for i,(m,n) in enumerate(par.items()):
    print(f'Kunder som har kjøpt: {m} Har også kjøpt {n[0]}, {n[1]}, {n[2]} og {n[3]}')
    if i == 4: break
```

```python
BASE = '../input/h-and-m-personalized-fashion-recommendations/images/'

for i,(m,n) in enumerate(par.items()):
    a1 = BASE+'0'+str(m)[:2]+'/0'+str(m)+'.jpg'
    a2 = BASE+'0'+str(n[0])[:2]+'/0'+str(n[0])+'.jpg'
    a3 = BASE+'0'+str(n[1])[:2]+'/0'+str(n[1])+'.jpg'
    a4 = BASE+'0'+str(n[2])[:2]+'/0'+str(n[2])+'.jpg'
    
    plt.figure(figsize=(20,5))
    img1 = cv2.imread(a1)[:,:,::-1]
    img2 = cv2.imread(a2)[:,:,::-1]
    img3 = cv2.imread(a3)[:,:,::-1]
    img4 = cv2.imread(a4)[:,:,::-1]
    plt.subplot(1,4,1)
    plt.title('Kunder som har kjøpt:',size=16)
    plt.imshow(img1)
    plt.subplot(1,4,2)
    plt.title('Har også kjøpt:',size=16)
    plt.imshow(img2)
    plt.subplot(1,4,3)
    plt.title('',size=16)
    plt.imshow(img3)
    plt.subplot(1,4,4)
    plt.title('',size=16)
    plt.imshow(img4)
    plt.show()
    if i == 4: break
```
